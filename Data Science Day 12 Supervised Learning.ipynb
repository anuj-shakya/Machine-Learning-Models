```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Linear Regression

Linear Regression can be computed either by using sklearn or scipy

```python
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error , r2_score
```

```python
X_train = np.random.normal(10,2,100)
X_test  = np.random.normal(10,2,30)
```

```python
y_train = X_train + np.random.normal(3,0.5,100)
y_test  = X_test  + np.random.normal(3,0.5,30)
```

```python
plt.scatter(X_train , y_train ,c = 'b')
plt.scatter(X_test , y_test , c = 'r')
plt.show()
           
```

```python
m, c, r2, p , err = linregress(X_train , y_train)
```

m = slope , c = y-intercept, r2 = correlation , p = pvalue , err = standard error (y - predicted Y)/n (sample)

```python
reg_line = m * X_train + c
```

```python
plt.scatter(X_train , y_train)
plt.plot(X_train , reg_line , c ='r')
plt.show()
```

```python
def predict(x_new):
    return m * x_new + c
```

```python
predict(12)
```

```python
# from sklearn method

lin_reg = LinearRegression(n_jobs = -1)
```

```python
X_train
```

```python
lin_reg.fit(X_train.reshape(-1,1) , y_train)
```

```python
lin_reg.intercept_
```

# return y - intercept $\theta_i$

```python
lin_reg.coef_
```

return slope - m =  $\theta_i$

```python
y_pred = lin_reg.predict(X_test.reshape(-1,1))
```

```python
y_test[0]
```

```python
y_pred[0]
```

```python
mean_absolute_error(y_test , y_pred)
```

```python
mean_squared_error(y_test , y_pred)
```

```python
r2_score(y_test ,y_pred)    
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,y_pred , c = 'r')
plt.show()
```
