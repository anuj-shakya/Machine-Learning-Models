```python
np.random.seed(666)

norm = np.random.normal(0,2,1000)
```

```python
plt.hist(x = norm , bins = 100)
plt.show
```

```python
norm.min()
```

```python
norm.max()
```

```python
# we had given Std = 2 at the begining to create nomral distribution

norm.std() 
```

```python
# variance = square of standard deviation

norm.var()
```

```python
# standard scaler

class StandardScaler:
    def __init__(self,):
        self.mean = None
        self.std  = None
        
    def fit(self,data):
        self.mean = data.mean()
        self.std = data.std()
        
    def transform(self,data):
        return ( data-self.mean ) / self.std
```

```python
scaler = StandardScaler()
scaler.fit(norm)
```

```python
scaled_norm=scaler.transform(norm)
scaled_norm
```

```python
# This is previous data 

norm
```

```python
plt.hist(x = norm, bins = 100)
plt.show
```

```python
plt.hist(x = scaled_norm, bins = 100)
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(low = 20 ,high = 100 , size = 15)
data
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, StandardScaler, RobustScaler
```

```python
names = ['MinMaxScaler','MaxAbsScaler','StandardScaler','RobustScaler']

scalers = {
    'MinMaxScaler' : MinMaxScaler(),
    'MaxAbsScaler' : MaxAbsScaler(),
    'StandardScaler': StandardScaler(),
    'RobustScaler' : RobustScaler()
}
```

```python
for scaler in scalers:
    print(f'while using :{scaler}')
    print(scalers[scaler].fit_transform(data.reshape(-1,1)))
    print('')
```
