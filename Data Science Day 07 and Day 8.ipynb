```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y1 = np.linspace(10,100,15)
y2 = y1 ** 2
y3 = y1 ** 3

x = np.linspace(1,15,15)
```

```python
# Numerical vs Numerical in line chart

plt.figure(figsize = (20,10))
plt.title('My plot')
plt.xlabel('Speed')
plt.ylabel('Distance')

plt.plot(x,y1, color = 'r', marker = 'X', label = 'Red')
plt.plot(x,y2, color = 'b', marker = '*', label = 'Blue')
plt.plot(x,y3, color = 'g', marker = '+', label = 'Green')

plt.legend()
plt.show()
```

```python
# pie chart ( Categorical Vs Numerical)

labels = [ 'A', 'B', 'C']
data   = [ 90, 40 , 60]
explode =[ 0, 0, 0.05]

plt.pie(x = data , labels = labels,explode = explode )
plt.show()
```
