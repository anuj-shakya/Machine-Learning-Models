```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Unsupervised Learning

```python
from sklearn.cluster import AgglomerativeClustering, KMeans
```

```python
data = pd.read_csv('../dataset/Mall_Customers.csv')
data.head()
```

```python
# X = data.iloc[:,3:]

X = data [['Annual Income (k)','Spending Score (1-100)']]
X.head()
```

```python
plt.scatter(x = X['Annual Income (k)'], y = X['Spending Score (1-100)'])
plt.show()
```

# KMeans Clustering

Sum of Squared Error

```python
sse = []     # sum of squared error

# more the sum of squared error, then it is kind of error
# less the sum of squared error, then it is right to accuracy

for clusters in range(1,10):
    k =KMeans(n_clusters = clusters)
    k.fit(X)
    sse.append(k.inertia_)
    
# inertia refers to sse
    
```

```python
plt.plot(range(1,10),sse)

plt.xlabel(xlabel= 'Clusters (K)')
plt.ylabel(ylabel= 'SSE (Sum of squared error)')
plt.title('Elbow Plot')
plt.show()
```

3 and 5 are the best model for clustering

```python
kmeans = KMeans(n_clusters = 5 , random_state = 666)
kmeans.fit(X)
```

```python
kmeans.labels_

# labelling our X data into 5 clusters ( 0,1,2,3,4)
```

```python
cluster_labels = kmeans.labels_
```

```python
X = X.values

# Converting dataframe into array frame
```

```python
plt.scatter(X[cluster_labels == 0,0],X[cluster_labels == 0,1], c = 'r')
plt.scatter(X[cluster_labels == 1,0],X[cluster_labels == 1,1], c = 'b')
plt.scatter(X[cluster_labels == 2,0],X[cluster_labels == 2,1], c = 'g')
plt.scatter(X[cluster_labels == 3,0],X[cluster_labels == 3,1], c = 'y')
plt.scatter(X[cluster_labels == 4,0],X[cluster_labels == 4,1], c = 'c')

plt.xlabel(xlabel= 'Annual Income')
plt.ylabel(ylabel= 'Spending Score')
plt.title('Annual Income VS Spending Score')

plt.show()
```

# AgglomerativeClustering

```python
agg = AgglomerativeClustering(n_clusters = 5)
agg.fit(X)
cluster_labels = agg.labels_
```

```python
plt.scatter(X[cluster_labels == 0,0],X[cluster_labels == 0,1], c = 'r')
plt.scatter(X[cluster_labels == 1,0],X[cluster_labels == 1,1], c = 'b')
plt.scatter(X[cluster_labels == 2,0],X[cluster_labels == 2,1], c = 'g')
plt.scatter(X[cluster_labels == 3,0],X[cluster_labels == 3,1], c = 'y')
plt.scatter(X[cluster_labels == 4,0],X[cluster_labels == 4,1], c = 'c')

plt.xlabel(xlabel= 'Annual Income')
plt.ylabel(ylabel= 'Spending Score')
plt.title('Annual Income VS Spending Score')

plt.show()
```
