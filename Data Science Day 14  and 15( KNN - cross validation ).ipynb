```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris_data = load_iris()
iris_data
```

```python
X = iris_data.data
y = iris_data.target
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size = 0.15, random_state= 666, stratify= y)
```

```python
knn_class = KNeighborsClassifier(n_jobs= -1, n_neighbors = 5)
```

```python
knn_class.fit(X_train, y_train)
```

```python
y_pred = knn_class.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test , y_pred)
```

```python
import seaborn as sns
sns.heatmap(confusion_matrix(y_test , y_pred),annot= True)
plt.show()
```

# Cross Validation

Grit search

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn_parameters = {
    'n_neighbors': [3,5,7,9,11],
    'weights':['uniform','distance']
}
```

```python
knn = KNeighborsClassifier(n_jobs=-1)
```

```python
grid_search = GridSearchCV( knn , param_grid= knn_parameters ,cv =10 , n_jobs= -1 )
```

```python
grid_search.fit(X_train , y_train)
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
best_knn = grid_search.best_estimator_
```

```python
best_knn.predict(X_test)
```
